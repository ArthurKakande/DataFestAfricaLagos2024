```python
"""
This function is used to clean the price of a mobile phone by removing commas and converting it to a float.#+
"""
def clean_price(price : str):
    price = re.sub(r"[,]", '', price)
    price = float(price)
    return price

data['Price'] = data['Price'].apply(clean_price)
data['Price'].head()
```

```python
# Name: 0, dtype: object
#Define your query
query = pd.DataFrame({
    "Spec_score" : [50],
    "No_of_sim" : ["Dual Sim"],
    "Ram" : ["8 GB RAM"],
    "Battery" : ["6000 mAh Battery"],
    "Display" : ["6.7 inches"],
    "Camera" : ["50 MP + 2 MP"],
    "External_Memory" : ["Memory Card Supported, upto 1 TB"],
    "Android_version" : ["13"],
    "Price" : [9999.0],
    "company" : ["Samsung"],
    "Inbuilt_memory" : ["128 GB"],
    "fast_charging" : ["25W Fast Charging"],
    "Screen_resolution" : ["2408 x 1080 px"],
    "Processor" : ["Octa Core Processor"],
})
```

```python
#Define you similarity calculation methods for your project
hamming = ik.sim_hamming
levenshtein = ik.sim_levenshtein
level = ik.sim_level
abs_diff = ik.sim_difference
```

```python
# Assign the appro*priate similarity calculation functions to each feature
similarity_functions = {
    "Spec_score" : abs_diff,
    "No_if_sim" : levenshtein,
    "Ram" : levenshtein,
    "Battery" : levenshtein,
    "Display" : levenshtein,
    "Camera" : levenshtein,
    "External_Memory" : levenshtein,
    "Android_version" : levenshtein,
    "Price" : abs_diff,
    "company" : hamming,
    "Inbuilt_memory" : levenshtein,
    "fast_charging" : levenshtein,
    "Screen_resolution" : levenshtein,
    "Processor" : levenshtein,
}
```

```python
# Applying the methods and weights and retrieving the top results using the linear retriever
feature_weights = {
    "Spec_score" : 0.2,
    "No_of_sim" : 0.2,
    "Ram" : 0.2,
    "Battery" : 0.3,
    "Display" : 0.1,
    "Camera" : 0.3,
    "External_Memory" : 0.1,
    "Android_version" : 0.3,
    "Price" : 0.5,
    "company" : 0.3,
    "Inbuilt_memory" : 0.1,
    "fast_charging" : 0.1,
    "Screen_resolution" : 0.1,
    "Processor" : 0.1,
}

top_n = 10  # Number of top similar results to return
top_similar_cases_linear = ik.linearRetriever(data, query, similarity_functions, feature_weights, top_n)
```

# Content Based Filtering

```python
metadata = pd.read_csv('movies_metadata.csv', low_memory = False)
metadata.head()
```

```python
metadata.columns
```

```python
metadata['overview'] = metadata['overview'].apply(lambda x: str(x))
```

```python
metadata[['original_title', 'overview']]
```

```python
metadata['overview'].loc[1]
```

```python
def get_content_based_recommendation(query, documents, n = 10):
    top_similar_docs_st = ik.vector_space_model(query, documents, k=n)

    similarity_dict = {
        'document' : [],
        'score' : [],
        'movie_titles' : []
    }

    for doc, similarity in top_similar_docs_st:
        similarity_dict['document'].append(doc)
        similarity_dict['score'].append(similarity)
        movie_title = metadata.loc[metadata['overview'] == doc]['original_title'].values[0]
        similarity_dict['movie_titles'].append(movie_title)


    return pd.DataFrame(similarity_dict).sort_values(by='score', ascending=False)
```

```python
query = "an action comedy movie"

documents = metadata['overview'].tolist()[:1000]
```

```python
get_content_based_recommendation(query, documents)
```
